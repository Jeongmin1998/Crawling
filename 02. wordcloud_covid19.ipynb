```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
corona_virus.head()
```

```python
comment_words = ''
stopwords = set(STOPWORDS)
```

```python
for val in corona_virus.USER_COMMENT:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    
    comment_words += " ".join(tokens)+" "
wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor =None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad =0)

plt.show()
```

```python
import re

comment_words = ''
stopwords = set(STOPWORDS)

for val in corona_virus.USER_COMMENT:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    
    comment_words += " ".join(tokens)+" "
    
comment_words = re.sub("[^a-zA-Z 0-9']",'', comment_words)
wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor =None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad =0)

plt.show()
```

## 이미지 모양 워드클라우드 만들기

```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

mask = np.array(cv2.imread('twitter.png')[:,:,::-1])
plt.imshow(mask)

wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10, mask=mask).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor =None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad =0)

plt.show()
```

```python
#이미지 사이즈 키워주면 됨


mask = np.array(cv2.imread('twitter.png')[:,:,::-1])
mask = cv2.resize(mask, (1000,1000))
plt.imshow(mask)

wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10, mask=mask).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor =None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad =0)

plt.show()
```
